```python
import pandas as pd  # Import the pandas library for data manipulation and analysis
import numpy as np   # Import the numpy library for numerical operations
import matplotlib.pyplot as plt  # Import the matplotlib library for data visualization
```

```python
df = pd.read_csv('E_commerce sample data.csv')  # Read the CSV file into a DataFrame named df
df  # Display the contents of the DataFrame
```

```python
df.head()  # Display the first five rows of the DataFrame df
```

```python
df.tail()  # Display the last five rows of the DataFrame df
```

```python
print(df.isnull().sum())  # Print the number of missing values in each column of the DataFrame df
df.describe().T  # Display summary statistics of the DataFrame df, transposed for better readability
```

```python
category_sales = df.groupby('ProductCategory')['OrderAmount'].agg(['sum', 'count']).reset_index()  
# Group the DataFrame df by 'ProductCategory' and aggregate 'OrderAmount' to calculate total sales and order count

category_sales.rename(columns={'sum': 'TotalSales', 'count': 'OrderCount'}, inplace=True)  
# Rename the columns for clarity: 'sum' to 'TotalSales' and 'count' to 'OrderCount'

category_sales  # Display the resulting DataFrame category_sales
```

```python
plt.figure(figsize=(10, 6))  # Create a new figure with a specified size
plt.bar(category_sales['ProductCategory'], category_sales['TotalSales'], color='black', alpha=0.8)  
# Create a bar plot with ProductCategory on the x-axis and TotalSales on the y-axis

plt.title('Total Sales by Product Category')  # Set the title of the plot
plt.xlabel('Product Category')  # Set the label for the x-axis
plt.ylabel('Total Sales')  # Set the label for the y-axis
plt.xticks(rotation=45)  # Rotate the x-axis labels by 45 degrees for better readability
plt.ylim(0, 4000)  # Set the limits for the y-axis from 0 to 4000
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add a grid to the y-axis with specified style and transparency
plt.tight_layout()  # Adjust the layout to make sure everything fits well
plt.show()  # Display the plot
```

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'])  
# Convert the 'OrderDate' column to datetime format

daily_orders = df.groupby(df['OrderDate'].dt.date)['OrderAmount'].agg(['count', 'mean']).reset_index()  
# Group the DataFrame by the date part of 'OrderDate' and aggregate 'OrderAmount' to get count and mean

daily_orders.rename(columns={'count': 'OrderCount', 'mean': 'AverageOrderValue'}, inplace=True)  
# Rename the columns for clarity: 'count' to 'OrderCount' and 'mean' to 'AverageOrderValue'

daily_orders  # Display the resulting DataFrame daily_orders
```

```python
plt.figure(figsize=(12, 6))  # Create a new figure with a specified size

plt.plot(daily_orders['OrderDate'], daily_orders['AverageOrderValue'], marker='o', color='Turquoise')  
# Plot AverageOrderValue against OrderDate with markers and a specified color

# plt.yticks(daily_orders['OrderDate'], rotation=45)  # (Commented out) This line would rotate y-ticks
# plt.xticks(daily_orders['AverageOrderValue'])  # (Commented out) This line would set x-ticks to AverageOrderValue

plt.xticks(rotation=45)  # Rotate the x-axis labels by 45 degrees for better readability
plt.grid(linestyle='--', alpha=0.7)  # Add a grid with a dashed line style to the plot
plt.ylim(0, 600)  # Set the limits for the y-axis from 0 to 600
plt.title('Daily Orders Count', fontsize=16)  # Set the title of the plot with a specified font size
plt.xlabel('Date', fontsize=12)  # Set the label for the x-axis with a specified font size
plt.ylabel('AverageOrderValue', fontsize=12)  # Set the label for the y-axis with a specified font size
plt.tight_layout()  # Adjust the layout to make sure everything fits well
plt.show()  # Display the plot
```

```python
top_ten_cust = df.groupby('CustomerName')['OrderAmount'].sum().nlargest(10).reset_index()  
# Group by 'CustomerName', sum the 'OrderAmount', get the top 10 customers by total order amount, and reset the index
top_ten_cust  # Display the resulting DataFrame
```

```python
plt.figure(figsize=(20, 10))  # Create a new figure with a specified size
color_cust = ['blue', 'green', 'red', 'purple', 'y', 'black', 'Turquoise', 'red', 'y', 'green']  
# Define a list of colors for the bars

plt.bar(top_ten_cust['CustomerName'], top_ten_cust['OrderAmount'], color=color_cust, label=top_ten_cust['CustomerName'])  
# Create a bar chart with CustomerName on the x-axis and OrderAmount on the y-axis

plt.grid(axis='y')  # Add a grid for the y-axis
plt.legend(loc='lower center', ncol=3, bbox_to_anchor=(0.5, -0.3), borderpad=1)  
# Add a legend at the specified location with multiple columns

plt.ylabel('OrderAmount')  # Set the label for the y-axis
plt.xlabel('CustomerName', fontsize=20)  # Set the label for the x-axis with a specified font size
plt.title('Top Ten Customers by Order Amount', fontsize=20)  # Set the title of the plot with a specified font size
plt.tight_layout()  # Adjust the layout to make sure everything fits well
plt.show()  # Display the plot
```

```python
order_status = df['OrderStatus'].value_counts()  
# Count the occurrences of each unique value in the 'OrderStatus' column
order_status  # Display the resulting Series
```

```python
plt.pie(order_status, 
        labels=order_status.index, 
        autopct='%1.1f%%', 
        startangle=140, 
        colors=plt.cm.Paired.colors)  
# Create a pie chart with the specified parameters
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.title('Order Status Distribution')  # Set the title of the pie chart
plt.show()  # Display the pie chart
plt.tight_layout()
```

```python
# Group by 'ProductCategory' to calculate the total revenue for each category
top_products = df.groupby('ProductCategory')['OrderAmount'].sum().nlargest(10).reset_index()

# Rename columns for better clarity
top_products.columns = ['ProductCategory', 'TotalRevenue']  

# Display the resulting DataFrame
top_products
```

```python
plt.figure(figsize=(12, 6))  
# Create a new figure with a specified size (width=12, height=6)

plt.bar(top_products['ProductCategory'], top_products['TotalRevenue'], color='purple', alpha=0.8)  
# Plot a bar chart with 'ProductName' on the x-axis and 'TotalRevenue' on the y-axis
# Use purple color for the bars and set their transparency to 0.8

plt.title('Top 10 Products by Revenue', fontsize=16)  
# Set the title of the chart with a font size of 16

plt.xlabel('Product Name', fontsize=12)  
# Label the x-axis as "Product Name" with a font size of 12

plt.ylabel('Total Revenue', fontsize=12)  
# Label the y-axis as "Total Revenue" with a font size of 12

plt.xticks(rotation=45, ha='right')  
# Rotate the x-axis labels by 45 degrees and align them to the right

plt.grid(axis='y', linestyle='--', alpha=0.7)  
# Add a dashed grid line along the y-axis with transparency set to 0.7

plt.tight_layout()  
# Automatically adjust the layout to ensure everything fits within the figure area

plt.show()  
# Display the plot
```

```python
df['OrderHour'] = df['OrderDate'].dt.hour  
# Extract the hour from the 'OrderDate' column and create a new column named 'OrderHour'

hourly_orders = df['OrderHour'].value_counts().sort_index().reset_index()  
# Count the number of orders for each hour, sort the index (hours) in ascending order, and reset the index for a clean DataFrame

hourly_orders.columns = ['OrderHour', 'OrderCount']  
# Rename the columns for clarity: 'index' becomes 'OrderHour', and 'OrderHour' becomes 'OrderCount'

hourly_orders  
# Display the resulting DataFrame, showing the number of orders for each hour
```

```python
plt.figure(figsize=(12, 6))  
# Create a new figure with a size of 12x6 inches

plt.plot(hourly_orders['OrderHour'], hourly_orders['OrderCount'], marker='o', color='blue', label='Orders')  
# Plot a line chart with 'OrderHour' on the x-axis and 'OrderCount' on the y-axis
# Add circular markers at each data point and set the line color to blue
# Add a label to the line for the legend

plt.title('Order Distribution by Hour of Day', fontsize=16)  
# Set the title of the chart with a font size of 16

plt.xlabel('Hour of Day', fontsize=12)  
# Label the x-axis as "Hour of Day" with a font size of 12

plt.ylabel('Number of Orders', fontsize=12)  
# Label the y-axis as "Number of Orders" with a font size of 12

plt.xticks(range(0, 24))  
# Set x-axis ticks to represent all 24 hours of the day (0 to 23)

plt.grid(axis='y', linestyle='--', alpha=0.7)  
# Add a dashed grid line along the y-axis with transparency set to 0.7

plt.tight_layout()  
# Automatically adjust the layout to ensure all elements fit well within the figure area

plt.legend()  
# Display the legend with the label specified earlier ('Orders')

plt.show()  
# Render and display the plot
```
